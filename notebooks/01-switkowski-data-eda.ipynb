### Product Name Length

The length of a product name is one of the first things a customer will see, and conveys the most important information: what the product is! There's a fine line between product names that are [too short] and [too long].

Which category the product falls under likely influences how long the product title should be. Let's take a deeper look into how long the product names are for different caregories.

First, I'll be looking at the **median** length of each product category.

```python
mid_prod_data = prod_data.groupby('product_category_name_english').agg(np.median).reset_index()
mid_prod_data.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

thousands_formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))

sns.barplot(
    y="product_category_name_english", x='product_name_lenght', 
    data=mid_prod_data.sort_values('product_name_lenght', ascending=[False]).reset_index(drop=True).iloc[:10], color='C0',
    ax=ax[0]
)

ax[0].set_title("10 Categories with Longest\n Median Product Name Length")
ax[0].set_ylabel("Product Category")
ax[0].set_xlabel("Count")
ax[0].get_xaxis().set_major_formatter(thousands_formatter)


sns.barplot(
    y="product_category_name_english", x='product_name_lenght', 
    data=mid_prod_data.sort_values('product_name_lenght', ascending=[False]).reset_index(drop=True).iloc[-10:], color='C0',
    ax=ax[1]
)

ax[1].set_title("10 Categories with Shortest\n Median Product Name Length")
ax[1].set_xlabel("Count")
ax[1].set_ylabel("")
ax[1].get_xaxis().set_major_formatter(thousands_formatter)
fig.tight_layout()
```

At the top of the median name length come a handful of home products, then some more "technical" categories like "Computers", "Tablets Printing Image", and "Auto". Of the other side, we see clothing fall towards the bottom, along with some other home goods like "Furniture" and "Christmas Supplies". 

I chuckle imagining some listings for a pair of shoes just listed as "Brown Shoes". In all seriousness though, the clothing category makes sense; pictures likely matter more than the *name* of the product. It's interesting to see Books fall towards the bottom here, although I suppose book titles can only be so long. Let's keep an eye on those categories, especially "Books Techincal" throughtout this analysis.

Next let's see how the overall distribution of the product name lenght is, not controlling for category.

```python
fig, ax = plt.subplots(figsize=(15, 7))
sns.histplot(prod_data['product_name_lenght'], ax=ax, kde=True, fill=False)
ax.axvline(prod_data['product_name_lenght'].median(), label='Median', color = 'C1', alpha = 0.75)
ax.axvline(prod_data['product_name_lenght'].mean(), label='Mean', color = 'C2', alpha = 0.75)

ax.set_xlabel('Product Name Length')
ax.set_title('Distribution of Product Name Length')
ax.legend();
```

This is a highly left-skewed distribution. Since the length takes a significant dip after 60, and shows no products with a length above 70, I suspect that there's a character or word limit on the product name put in place by Olist.

Let's see if this distribution is shared by individual categories.

```python
def plot_dist(ax, category, column):
    
    cat_data = prod_data[prod_data['product_category_name_english'] == category]
    sns.histplot(cat_data[column], ax=ax, kde=True, fill=False)
    ax.axvline(cat_data[column].median(), label='Median', color = 'C1', alpha = 0.75)
    ax.axvline(cat_data[column].mean(), label='Mean', color = 'C2', alpha = 0.75)

    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title(category.replace('_', ' ').title(), fontsize=8)

fig, ax = plt.subplots(9, 8, figsize=(20, 20))

for idx, product in enumerate(mid_prod_data['product_category_name_english'].tolist()):
    i = int(np.floor(idx / 8))
    j = idx%8
    plot_dist(ax[i][j], product, 'product_name_lenght')
fig.tight_layout()
```

Most of the categories, especially the ones with a large number of products, show the same distribution. Some categories have barely any products show odd distributions that we can ignore.

However, there are some that are show a normal distribution, especially "Perfumery". I wonder if this is because of another "limit" on the length of the product name since it's often printed on the perfume bottle. "Office furniture", "Furniture Living Room", and all the "Books" categories are some other categories that break from the population's distribution.

### Product Description Length

Next, let's look at the product description length. This contains specific information about the product that the customer may need to know before buying. Again, we'll look at the **median** length for the product description.

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

thousands_formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))

sns.barplot(
    y="product_category_name_english", x='product_description_lenght', 
    data=mid_prod_data.sort_values('product_description_lenght', ascending=[False]).reset_index(drop=True).iloc[:10], color='C0',
    ax=ax[0]
)

ax[0].set_title("10 Categories with Longest\n Median Product Name Length")
ax[0].set_ylabel("Product Category")
ax[0].set_xlabel("Length")
ax[0].get_xaxis().set_major_formatter(thousands_formatter)


sns.barplot(
    y="product_category_name_english", x='product_description_lenght', 
    data=mid_prod_data.sort_values('product_description_lenght', ascending=[False]).reset_index(drop=True).iloc[-10:], color='C0',
    ax=ax[1]
)

ax[1].set_title("10 Categories with Shortest\n Median Product Name Length")
ax[1].set_xlabel("Length")
ax[1].set_ylabel("")
ax[1].get_xaxis().set_major_formatter(thousands_formatter)
fig.tight_layout()
```

At the number one spot lands "Comptuers". This makes a lot of sense to me since the photos and description can't accurately describe what a custoemr is purchasing. When considering hardware, specs, and supported devices, and then add in all teh SEO keywors, it's easy to see why it's at the top. 

Following "Computers" is "Books Technical". Again, this makes a ton of sense; there needs to be a verbose description of what the book is about when listing it on an ecommerce website. If nothing else, the seller is likely copying the commentary in the inner folds and the back of the book to the description.

As an aside, this would be a great ML application: populating the description of a product based on the name of the product, especially for books.

Let's now look at the distribution of the product description like we did with the name length.

```python
fig, ax = plt.subplots(figsize=(15, 7))

sns.histplot(prod_data['product_description_lenght'].values, kde=True, fill=False)
ax.axvline(prod_data['product_description_lenght'].median(), label='Median', color = 'C1', alpha = 0.75)
ax.axvline(prod_data['product_description_lenght'].mean(), label='Mean', color = 'C2', alpha = 0.75)

ax.set_xlabel('Product Name Length')
ax.set_title('Distribution of Product Name Length')
ax.legend();
```

Unlike the product name, this distribution is highly right skewed. While there is probably a cap at 4,000 characters it's difficult for most sellers to hit that limit. 

```python
fig, ax = plt.subplots(9, 8, figsize=(20, 20))

for idx, product in enumerate(mid_prod_data['product_category_name_english'].tolist()):
    i = int(np.floor(idx / 8))
    j = idx%8
    plot_dist(ax[i][j], product, 'product_description_lenght')
fig.tight_layout()
```

```python
sns.jointplot('product_description_lenght', 'product_name_lenght', data=prod_data, height=10, kind='kde')
```
