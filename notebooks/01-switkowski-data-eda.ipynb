Where are most of the customer's from?

```python
fig, ax = plt.subplots(1, 2, figsize=(7*3, 7))

thousands_formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))

sns.countplot(
    y="customer_state", data=cust_data, color='C0', 
    order=cust_data['customer_state'].value_counts().index,
    ax=ax[0]
)

ax[0].set_title("Customer State Count")
ax[0].set_ylabel("Customer State")
ax[0].set_xlabel("Count")
ax[0].get_xaxis().set_major_formatter(thousands_formatter)

map_count = map_data[['sigla', 'geometry']]\
.merge(
    cust_data['customer_state']
    .value_counts()
    .reset_index()
    .rename(columns={'index':'state', 'customer_state':'count'}), 
    left_on='sigla', right_on='state')\
.drop('sigla', axis=1)

map_count.plot(
    column='count',
    cmap='summer', 
    legend=True,
    ax=ax[1]);
ax[1].set_xticks([])
ax[1].set_yticks([]);
```

How about the customer city? 

```python
map_data = gpd.read_file('../data/raw/bcim_2016_21_11_2018.gpkg', layer='lim_municipio_a')
map_data.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(7*3, 7))

thousands_formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))

sns.countplot(
    y="customer_city", data=cust_data, color='C0', 
    order=cust_data['customer_city'].value_counts().iloc[:25].index,
    ax=ax[0]
)

ax[0].set_title("Customer City Count")
ax[0].set_ylabel("Customer State")
ax[0].set_xlabel("Count")
ax[0].get_xaxis().set_major_formatter(thousands_formatter)

map_count = map_data[['nome', 'geometry']]\
.merge(
    cust_data['customer_city']
    .value_counts()
    .reset_index()
    .rename(columns={'index':'city', 'customer_city':'count'}), 
    left_on='nome', right_on='city', how='left')\
.fillna(0)\
.drop('nome', axis=1)

map_count.plot(
    column='count',
    cmap='summer', 
    legend=True,
    ax=ax[1]);
ax[1].set_xticks([])
ax[1].set_yticks([]);
```
